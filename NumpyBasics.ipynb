```python
arrh = np.empty(1000)

for i in range(1000):
    cf = np.random.randint(0, 2, 1000)
    cfh = np.sum(cf)
#    arrh = np.append(arrh, cfh)
    arrh[i] = cfh
    
sum = np.sum(arrh)
print('Sum: ' + str(sum))
avg = np.average(arrh)
print("Mean: " + str(avg))
var = np.var(arrh)
print("Variance: " + str(var))

xval = np.arange(1000)
plt.figure(figsize=(15,5))
plt.plot(xval, arrh, xval, np.repeat(avg, 1000), "r-")
plt.show
```
