```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_line(m, b, t0, tf, dt):    
    x = np.arange(t0, tf, dt)
    y = (x*m)+b
    return x,y
    
x,y = gen_line(4,12,0,50,10)
plt.plot(x, y, linestyle='-', marker='o', color='k')
plt.show()
```

```python
def gen_line_stoch(m, b, t0, tf, dt, var=1):
    x = np.arange(t0, tf, dt)
    y = (x*m)+b
    z = np.random.normal(y, var)
    return x,z
    
x,z = gen_line_stoch(2,0,0,40,1,3)
plt.plot(x, z, linestyle='-', marker='o', color='k')
plt.show()
```

# Linear Regression

We will do linear regression first with projection and then with gradient descent.

### Linear Regression with Projection

We find the slope and y-intercept of the line of regression by projecting $z$, or all the given data output points, onto the space spanned by $x$, the input points. This is because the span of $x$ consists of $mx$ for some $m$, which we are considering our 'slope'. The projection will give a series of outputs that are the closest to $z$ as possible, while still being expressed by a line $z_{proj} = mx + b$. 

In order to express the elements of the projection as matrices and vectors, and solve for the slope and y-intercept, we create the following:

$M$,a matrix consisting of a column of 1s and a column with the values of x (the input values).
$$
M = 
\begin{bmatrix}
1 & x_1 \\
1 & x_2 \\
\vdots & \vdots \\
1 & x_n
\end{bmatrix}
$$
$A$, a two-element vector containing $m_1$ and $b_1$, the representation of slope and y-intercept for the line of best fit that we want to find. 
$$
A = 
\begin{bmatrix}
m_1 \\ b_1
\end{bmatrix}
$$
When this vector $A$ is multiplied by the matrix $M$, the resulting vector will be the points correlating to the $x$-inputs, on the line of regression. <br>

To find $m_1$ and $b_1$ such that the line of regression is closest to the outputs we have (populated in $z$), we solve $MA=z$. Since M is not invertible we multiply both sides by $M^T$, to force $M^T M$ to be invertible. Thus we achieve the formula

$$
A = (M^T  M)^{-1}  M^T  z
$$

This is implemented below to return $m_1$ and $b_1$.

```python
def linreg_proj(x, z):
    M = np.ones([len(x), 2])
    M[:, 1] = x

    A = np.linalg.inv(np.transpose(M) @ M) @ np.transpose(M) @ z
    b1 = A[0]
    m1 = A[1]
    
    return m1, b1


m1, b1 = linreg_proj(x,z)
plt.scatter(x, z, marker='o', color='k')
y1 = m1*x+b1
plt.plot(x, y1, linestyle='--', color='b')
plt.show()
```

### Linear Regression with Gradient Descent

We build an iterative model that arrives at slope $m$ and y-intercept $b$ by minimizing the difference between the data output value (in $z$) and the predicted values (given by $P = mx + b$).

The loss function collects the model's error by squaring the difference between the model's output and the real output (to keep it positive) and finding the average of the error of the data set: 

$$
L = \frac{1}{n} \sum_{i=0}^{n} (z_i - P_i)^2
$$

$$
L = \frac{1}{n} \sum_{i=0}^{n} (z_i - (m x_i + b))^2
$$

When this average error is minimized, we have achieved the closest line to the data set that is possible. 

The model learns by moving $m$ and $b$ by increments of $d$, which we will set to $0.0001$. The model will be most accurate when $d$ is small, so that $m$ and $b$ are changed slowly.

We change $m$ and $b$ by monitoring the changes in the loss function with respect to each variable. The partial derivatives are given below:

$$
L_m = \frac{1}{n} \sum_{i=0}^{n} 2(z - (m x_i + b))(-x_i) = \frac{-2}{n} \sum_{i=0}^{n} (x_i)(z - mx_i - b)
$$

$$
L_b = \frac{-2}{n} \sum_{i=0}^{n} (z - (mx_i + b))
$$

With each iteration, we calculate the partial derivatives and adjust $m$ and $b$ accordingly:

$$ 
m = m + d * L_m
$$
$$
b = b + d * L_b
$$

After all the iterations have been run, the values of $m$ and $b$ will be close to that of the line of best fit.

```python
def linreg_grad(x, z, iter=1000, d=0.0001):
    # Set variables
    m = 0
    b = 0
    n = len(x)

    for i in range(n):
        P = m*x + b
        L_m = (-2/n) * sum(x*(z - P))
        L_b = (-2/n) * sum(z - P)
        m = m - d*L_m
        b = b - d*L_b
    
    return(m,b)
    
m,b = linreg_grad(x,z)
P = m*x + b
print(m,b)
plt.scatter(x,z, color='k')
plt.plot(x,P, linestyle='--', color='r')
plt.show
```

# K-Means Clusters

## Data Simulation
We define $k$ to be the number of clusters that data will be generated around. The value of each cluster's mean is determined by a $var\_random$, a random variable used to generate the values of each mean (the kernel value we will use is 0). The third input is $var\_cluster$: if it is a list, it must contain $k$ elements and represent the variances for the data points generated in each cluster. If it is a single value, it will represent a variance used for generating data points in all clusters.

## Clustering Algorithm
The clustering algorithm takes in the two lists (of $x$-coordinates and $y$-coordinates) and uses a loop of three steps to identify clusters of points and subsequently what the means of those clusters are.



```python
def k_simulator(k, var_random, var_cluster):
    
    # Generating means
    means = np.ndarray((k, 2))
    for i in range(k):
        for j in range(2):
            means[i][j] = np.random.normal(0, var_random)
    plt.scatter(means[:,0], means[:,1], color='k')
    
    
    # Generating points for each cluster
    # Using a seed value of 10 and a variance of 1 to determine how many points total are generated
    xpoints = np.zeros(int(k*np.random.normal(10, 1)))
    ypoints = np.zeros(len(xpoints))

    # Data point generation uses a list for cluster variance, so if passed in value is not a list, create one
    if isinstance(var_cluster, list):
        pass
    else: 
        var_temp = var_cluster
        var_cluster = np.repeat(var_temp, k)
    split = int(len(ypoints)/k)
    
    # Generate data points by looping through each section of the xpoints array and inserting randomized values
    for i in range(k):
        for j in range(split):
            xpoints[(i*split)+(j)] = np.random.normal(means[i-1, 0], var_cluster[i])
            ypoints[(i*split)+(j)] = np.random.normal(means[i-1, 1], var_cluster[i])
            
    # Due to split algorithm, sometimes the last few values don't get filled; adjust for those.
    if len(xpoints) != k*split:
        xpoints = xpoints[0:k*split]
        ypoints = ypoints[0:k*split]
        
    plt.scatter(xpoints, ypoints, color='y')    

        
k_simulator(6, 2, 0.2)
```
